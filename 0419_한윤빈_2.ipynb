```python
# 점수검증함수

def correct(inputnum):
  A = inputnum()
  error = []
  for i in range(6):
   for j in range(5):
     for k in range(5):
       if  A[i][j][k]< 0 or A[i][j][k] > 100:
         print(i,j,k,A[i][j][k], "은 잘못된 값")
         error.append([i,j,k])
     
  return error
```

```python
# 반 별 평균 리스트
def average(inputnum):
  A = inputnum()
  average = []
  count = 0
  sum = 0

  for i in range(len(A)):     
    for j in range(len(A[i])):     
      for k in range(len(A[i][j])): 
        sum += A[i][j][k]
      average.append(sum/5)
      sum = 0
  return average # 반 별 평균 리스트
```
